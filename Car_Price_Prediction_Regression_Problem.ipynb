```python
# Filling Missing Values
for value in ['Engine', 'Max Power','Max Torque', 'Drivetrain']:
    data[value].fillna(data[value].mode()[0], inplace=True)

for value in ['Length', 'Width', 'Height','Seating Capacity', 'Fuel Tank Capacity']:
    data[value].fillna(data[value].median(), inplace=True)
```

### Assumption 3: Assumption of Normality: 
The dependet variable should follow an approximate normal Distribution

```python
sns.distplot(y);
```

**Observation:**
Price variable is not meeting the assumption and we can see that y variable is right skewed and not following proper normal Distribution

**Note: If its not normal or if its skewed plot then we can transform the column in different transformation**
1. Log Transform : Most of the time we prefer log transform, because it will help you to smoothing of curve
2. SQRT Transform: In sqrt transform we can see the peeks and sometime it will not help you to smoothing of curve

```python
y_log = np.log1p(y)
sns.distplot(y_log)
```

**Observation:** If your data is approximate normal then dont do transform because you will get skewed data. In our case we will go ahead with y values intead of taking y_log values since after transformation the data is more skewed towards left.

**Note: For good practive make sure that your X variables does not contain high skewness and if its approximately normal then its good.**

```python
#checking for skewness for x variable
x.hist(bins=20);
```

**Observation:** From above histograms we can clearly see year is left skewed and length, width and fuel tank capacity are little right skewed where as other variables follow normal distribution curve.

```python
#checking for skewness of all x variables 
from scipy.stats import skew
#finding the skewness in each variable
data_num_skew = x.apply(lambda i: skew(i.dropna()))
#subsetting the variables that are highly skewed
data_num_skewed = data_num_skew[(data_num_skew > .75) | (data_num_skew < -.75)]
 
print(data_num_skew)
print(data_num_skewed)

# apply log + 1 transformation for all numeric features with skewness over .75
x[data_num_skewed.index] = np.log1p(x[data_num_skewed.index])
```

**Note:** Range of skewness is -1 to +1 and value close to 0 means normal distribution

### Assumption 4:- Assumption of no multicollinearity:
There should be No multicolinearity in the Data. 

Multicolinearity is problem where one variable is dependent on other, we want dependecy between x and y variable but not in the x variables, so if x variables are depend on each other then the model will not be good.

To check the multicolinearity we can use the pearson correlation or VIF value

```python
# Pearson Correlation
corr_df=x.corr(method="pearson")
print(corr_df)
 
sns.heatmap(corr_df,vmax=1.0,vmin=-1.0,annot=True)
plt.show()
```

**Observation:** We want value close to 0 which means their is no correlated with each other.

**Note: If the correlation is high between two variables then drop one beacuse bothe varible carrying the similar information so eliminate the varible but think logically.**

**Check for VIF value:** Sometimes the correlation matrix alone will not help you to check the multicolinearity so you can use VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

vif_df = pd.DataFrame()
vif_df["features"] = x.columns
vif_df["VIF Factor"] = [vif(x.values, i) for i in range(x.shape[1])]
vif_df.round(2)
```

**Observation:** VIF value Less than 5 is good Which represent no multicolinearity

**Note:** Once you eliminate one variable based on VIF score then other variable's VIF also change and it will decrease so never eliminate all variables at once, so remove variables with highest value and then run the function again. Also use domain knowledge dont just depend on VIF values.

## Splitting the Data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.3, random_state=10)

# > 1000 --> TEST SIZE = 0.3
# < 1000 --> TEST SIZE = 0.2
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

## Scalling the Data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)

x_train_scale = scaler.transform(x_train)  
x_test_scale = scaler.transform(x_test)  
```

```python
x_train
```

```python
x_train_scale
```

## Model Building using Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm=LinearRegression()

# train the model object
lm.fit(x_train_scale, y_train)  
```

```python
# print intercept and coefficients
print (lm.intercept_)
print (lm.coef_)
```

```python
# pair the feature names with the coefficients
print(list(zip(x.columns, lm.coef_)))
```

```python
y_pred=lm.predict(x_test_scale)
print(y_pred) 
```

```python
new_df=pd.DataFrame()
new_df=x_test.copy()
 
new_df["Actual Price"]=y_test
new_df["Predicted Price"]=y_pred
new_df  
```

```python
from sklearn.metrics import r2_score,mean_squared_error

print("R-squared:", r2_score(y_test,y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test,y_pred)))
print("Adj R-square:", 1 - (1-r2_score(y_test,y_pred))*(len(y)-1)/(len(y)-x.shape[1]-1))
```

**Observation:** R-square value tells you how much variablity you can explaoin in the data with the help of regression equation. You will get value in the range of 0 to 1 so close to 1 means good model and close to 0 means worst model (sometimes you will get value less than 0 also)

```python
new_df["Error"] = new_df["Actual Price"] - new_df["Predicted Price"]
new_df.to_csv("Car_Price_Pred.csv", header=True, index=False)
new_df.head()
```
